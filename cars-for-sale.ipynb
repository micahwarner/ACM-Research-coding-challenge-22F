```python
#Basic imports
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
def check(list, compare):
    for x in list:
        if x == compare:
            return False
    return True

def getNewest(arr, compare):
    i = 0;
    x = len(arr);
    max = arr[i][0];
    while i < x:
        if(compare == arr[i][2] and arr[i][0] > max):
            max = arr[i][0]
        i+=1
    return max
    
def getLowMiles(arr, compare):
    i = 0;
    x = len(arr)
    min = arr[i][31]
    while i < x:
        type = arr[i][2]
        if (type == compare):
            if(min > arr[i][31]):
                min = arr[i][31]
        i+=1
    return (min)

def plotColors(arr):
    figure(figsize=(8, 6), dpi = 1000)
    plt.rc('xtick', labelsize=6) 
    i = 0;
    x = 50;
    colors = [];
    colorCount = [];
    while i < x:
        tempColor = arr[random.randint(1, len(arr))][21];
        if(not check(colors, tempColor)):
            colorCount[colors.index(tempColor)]+=1;
        else:
            colors.append(tempColor);
            colorCount.append(1);
        i+=1;
    plt.bar(colors, colorCount)
    plt.xlabel('Color')
    plt.ylabel('Amount')
    plt.xticks(rotation = "vertical")    
    plt.show()

def plotTypes(arr):
    figure(figsize=(8, 6), dpi = 1000)
    plt.rc('xtick', labelsize=6) 
    i = 0;
    x = 50;
    types = [];
    typeCount = [];
    while i < x:
        tempType = arr[random.randint(1, len(arr))][2];
        if(not check(types, tempType)):
            typeCount[types.index(tempType)]+=1;
        else:
            types.append(tempType);
            typeCount.append(1);
        i+=1;
    plt.bar(types, typeCount)
    plt.xlabel('Type')
    plt.ylabel('Amount')
    plt.xticks(rotation = "vertical")    
    plt.show()

df = pd.read_csv('../input/carsforsalecode/cars_raw.csv')
arr = df.to_numpy()
carTypes = []
plotColors(arr);
plotTypes(arr);
for type in arr[:, 2]:
    carTypes.append(type)
typesNoRepeat = []
typeCount = []
i = 0
n = 0
x = len(carTypes)
while i < x:
    currentType = carTypes[i]
    if check(typesNoRepeat,currentType):
        typesNoRepeat.append(currentType)
        typeCount.append(1)
    else:
        typeCount[typesNoRepeat.index(currentType)]+=1
    i+=1
while n < len(typesNoRepeat):
    print(typesNoRepeat[n] + " Amount: " + str(typeCount[n]) +" Lowest Mileage: " + str(getLowMiles(arr, typesNoRepeat[n])) + " Newest: " + str(getNewest(arr, typesNoRepeat[n])))
    n+=1
```
